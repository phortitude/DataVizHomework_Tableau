# COMBINE TRIP DATA FILES 

```python
# path = r'C:\DRO\DCL_rawdata_files' # use your path
all_files = glob.glob("*.csv")



trip_files = glob.glob("*tripdata*")
rider_files = glob.glob("*rider*")
li = []
    
print(trip_files)
print(rider_files)

for filename in all_files:
    all_tripdata_df = pd.read_csv(filename, index_col=None, header=0)
    li.append(all_tripdata_df)

trips_2018_2019_df = pd.concat(li, axis=0, ignore_index=True)
```

```python
trips_2018_2019_df
```

```python
# save combined trip data files to csv
trips_2018_2019_df.to_csv('all_trips_2018_2019')
```
